```python
from esperanto import OpenAILanguageModel, AnthropicLanguageModel, GeminiLanguageModel, VertexAILanguageModel, GroqLanguageModel, VertexAnthropicLanguageModel, OllamaLanguageModel, XAILanguageModel, OpenRouterLanguageModel, LiteLLMLanguageModel
input_text = "Why is the sky blue?"
```

```python
ProviderModel = [
    {"provider": OpenAILanguageModel, "model_name": "gpt-3.5-turbo"},
    {"provider": AnthropicLanguageModel, "model_name": "claude-3-sonnet-20240229"},
    {"provider": GeminiLanguageModel, "model_name": "gemini-pro"},
    {"provider": VertexAILanguageModel, "model_name": "gemini-pro"},
    {"provider": GroqLanguageModel, "model_name": "llama-3.1-8b-instant"},
    {"provider": VertexAnthropicLanguageModel, "model_name": "claude-3-5-sonnet@20240620"},
    {"provider": OllamaLanguageModel, "model_name": "gemma2"},
    {"provider": XAILanguageModel, "model_name": "grok-beta"},
    {"provider": LiteLLMLanguageModel, "model_name": "openai/gpt-4o-mini"},
    {"provider": OpenRouterLanguageModel, "model_name": "infermatic/mn-inferor-12b"}
    ]
```

```python
for provider_model in ProviderModel:
    try:
        provider = provider_model["provider"]
        model_name = provider_model["model_name"]
        print("Provider:", provider_model["provider"])
        print("Model name:", model_name)
        llm = provider(model_name=model_name)
        langchain_model = llm.to_langchain()
        response = langchain_model.invoke(input_text)
        print("Response: OK")
    except Exception as e:
        print("Failed to execute:", str(e))
    finally:
        print("------------------------")
```
