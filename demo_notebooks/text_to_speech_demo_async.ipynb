# Text-to-Speech Demo (Async)

This notebook demonstrates how to use different text-to-speech providers in Esperanto with async/await support.

We'll use the following providers:
1. OpenAI TTS
2. Gemini TTS
3. ElevenLabs TTS

First, let's import the necessary modules and set up our providers:

```python
import os
import asyncio
from pydantic import SecretStr
from esperanto.providers.text_to_speech.openai import OpenAITextToSpeechModel
from esperanto.providers.text_to_speech.gemini import GeminiTextToSpeechModel
from esperanto.providers.text_to_speech.elevenlabs import ElevenLabsTextToSpeechModel

# Helper function to play audio
from IPython.display import Audio, display

def play_audio(audio_file):
    """Play audio from a file using IPython's Audio display."""
    display(Audio(filename=audio_file))
```

## 1. OpenAI Text-to-Speech

Let's try OpenAI's text-to-speech service first. They offer multiple voices and models:

```python
# Initialize OpenAI TTS model
openai_tts = OpenAITextToSpeechModel(
    model_name="tts-1",
    config={
        "voice": "alloy",  # Available voices: alloy, echo, fable, onyx, nova, shimmer
        "response_format": "mp3"
    }
)

# Text to synthesize
text = "Hello! This is a test of the OpenAI text-to-speech system."

# Create output directory if it doesn't exist
os.makedirs("output", exist_ok=True)

# Synthesize and play audio
output_file = "output/openai_test.mp3"
await openai_tts.synthesize(text, output_file)
play_audio(output_file)
```

## 2. Gemini Text-to-Speech

Now let's try Google's Gemini text-to-speech service:

```python
# Initialize Gemini TTS model
gemini_tts = GeminiTextToSpeechModel(
    model_name="gemini-pro",
    config={
        "language_code": "en-US",
        "name": "en-US-Neural2-C",  # Voice name
        "speaking_rate": 1.0,
        "pitch": 0.0
    }
)

# Text to synthesize
text = "Hello! This is a test of the Gemini text-to-speech system."

# Synthesize and play audio
output_file = "output/gemini_test.mp3"
await gemini_tts.synthesize(text, output_file)
play_audio(output_file)
```

## 3. ElevenLabs Text-to-Speech

Finally, let's try ElevenLabs' text-to-speech service:

```python
# Initialize ElevenLabs TTS model
elevenlabs_tts = ElevenLabsTextToSpeechModel(
    model_name="eleven_multilingual_v2",
    config={
        "voice_id": "21m00Tcm4TlvDq8ikWAM",  # Rachel voice
        "optimize_streaming_latency": 0,
        "output_format": "mp3"
    }
)

# Text to synthesize
text = "Hello! This is a test of the ElevenLabs text-to-speech system."

# Synthesize and play audio
output_file = "output/elevenlabs_test.mp3"
await elevenlabs_tts.synthesize(text, output_file)
play_audio(output_file)
```
