```python
from esperanto.providers.embedding import OpenAIEmbeddingModel, VertexEmbeddingModel, GeminiEmbeddingModel, OllamaEmbeddingModel, OllamaEmbeddingModel
texts = ["This is a test", "Another test"]
```

## OpenAI Embedding Model

```python
model = OpenAIEmbeddingModel(model_name="text-embedding-3-small")
embeddings = await model.embed(texts)
print(embeddings)
```

## Vertex Embedding Model

```python
vertex_model = VertexEmbeddingModel(model_name="textembedding-gecko-multilingual@001")
embeddings = await vertex_model.embed(texts)
print(embeddings)
```

## Gemini Embedding Model

```python
gemini_model = GeminiEmbeddingModel(model_name="models/text-embedding-004")
embeddings = await gemini_model.embed(texts)
print(embeddings)
```

## Ollama Embedding Model

```python
ollama_model = OllamaEmbeddingModel(model_name="mxbai-embed-large")
embeddings = await ollama_model.embed(texts)
print(embeddings)
```
