```python
from esperanto import OpenAISpeechToTextModel, GroqSpeechToTextModel

audio_file = 'test_files/audio.mp3'
```

## OpenAI Speech-to-Text Model


```python
openai_model = OpenAISpeechToTextModel(model_name='whisper-1')
result = await openai_model.transcribe(audio_file)
print(result)
```

## Groq Speech-to-Text Model

```python
groq_model = GroqSpeechToTextModel(model_name='whisper-large-v3-turbo')
result = await groq_model.transcribe(audio_file)
print(result)
```
